## Business Understanding

In this lab we are exploring the [Reuters 50 50](https://archive.ics.uci.edu/ml/datasets/Reuter_50_50) Dataset from the UCI Repository. The dataset includes a training set and a test set, each containing 50 works of 50 different authors, for a total of 2,500 articles. This dataset allows us to explore the field of authorship identification, and see how different authors write. According to UCI, the articles chosen are under the "corporate/industrial" subtopic to reduce the ability to distinguish through topics alone. However, since the topic is still very broad and authors may cover different regions of the world, we will see if some words appear under one author more than others. We will only be visualizing the training set (C50training) in this lab, but the dataset will save us future steps in testing classification models later on. Furthermore, we may use this type of dataset to predict biases or sentiments among different authors -- i.e. is one author more likely to cover one region more positively than another?

```python
import pandas as pd
import numpy as np
import os
from scipy.misc import imread
import matplotlib.pyplot as plt
import random
import warnings
import seaborn as sns


directory = os.path.expanduser('~/Downloads/C50/C50train/')
articles = []
authors = []

# loop over directories: 
for author in os.listdir(directory):
    folder = os.path.join(directory, author)
    
    # loop over files in folder
    for filename in os.listdir(folder):
        with open(os.path.join(folder, filename)) as f:
            articles.append(f.read()) # append to array
            authors.append(author)
```

## Data Encoding

To ensure data quality, the following code cleans our data of any numbers or non-english characters that do not care about in our data set. Then we de-tokenize the words and get rid of any stop words that appear in the text. We also chose not to stem words, as certain authors may use a certain form of the word more than others. For instance, some authors may write in the present tense and others in the past tense. Stemming words would match different tenses to the same word, which may alter some predictions later on. Our cleaning process eliminates approximately 7% of the data set, or 1,794 words. 

```python
import re
from nltk.corpus import stopwords
cachedStopWords = stopwords.words('english')

cleaned_articles = []
for item in articles:
    letters = re.sub("[^a-zA-Z]", " ", item)
    lower_case = letters.lower()
    text = ' '.join([word for word in lower_case.split() if word not in cachedStopWords])
    cleaned_articles.append(text)
```

### Bag of Words

We create a "bag of words" using scikit learn to determine how frequently each word appear in the corpus. There are 27,343 unique words in the dataset, and our sparse matrix counts how many times each word appears in each of the 2,500 articles. We sum up each column to find the number of times a word appears in the entire dataset, and sort by the most commonly used words. We later repeat the process for each author's group of articles. 

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()

bag_words = count_vect.fit_transform(cleaned_articles)
```

```python
print(bag_words.shape) # this is a sparse matrix
print('=========')
# print(bag_words[0])
```

```python
pd.options.display.max_columns = 999
df_frequency = pd.DataFrame(data=bag_words.toarray(),columns=count_vect.get_feature_names())

df_frequency.head()
```

```python
print(len(count_vect.vocabulary_))
```

### World Relevance: TF-IDF

Here we create a sparse encoded term frequency, inverse document frequency (TF-IDF) representation of the word corpus. TF-IDF normalizes word frequency to create a measure of importance, so we can use this representation to find the most prominent/relevant words in the entire corpus. We can also apply TF-IDF to an individual author's set of texts for comparison with other authors. Perhaps there are certain words or terms more important to one author as opposed to another, or to the corpus as a whole. 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect = TfidfVectorizer()

tfidf_mat = tfidf_vect.fit_transform(cleaned_articles) 
```

```python
# convert to pandas to get better idea about the data
df = pd.DataFrame(data=tfidf_mat.toarray(),columns=tfidf_vect.get_feature_names())
df.head()
```

The following graph shows a seaborn kernel density plot of each author's top 20 word frequencies -- computed by the sum of each authors columns in the bag of words model. By visualizing the data in this manner, we can attempt to differentiate each author's writing styles through the distribution of frequency of words. Unfortunately, this graph becomes really cluttered at 50 different authors and is hard to differentiate between the authors, so we only examine 5 authors. If we compared authors a few at a time, we can see slight differences in how often an author likes to repeat his or herwords. 

```python
import itertools

author_names = ['AaronPressman', 'AlanCrosby', 'AlexanderSmith', 'BenjaminKangLim', 'BernardHickey']
#                 'BradDorfman', 'DarrenSchuettler', 'DavidLawder', 'EdnaFernandes', 'EricAuchard']
# for author in os.listdir(directory):
#     author_names.append(author)

ax = sns.set_style('whitegrid')
palette = itertools.cycle(sns.color_palette())

plt.figure(figsize=(10,10))    
for author in author_names:
    bag_words = count_vect.fit_transform(by_author[author])
    df_frequency = pd.DataFrame(data=bag_words.toarray(),columns=count_vect.get_feature_names())
    freq = df_frequency.sum().sort_values()[-20:]
    sns.kdeplot(np.array(freq), color=next(palette), shade=True, label=author)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Authors Top Word Frequency Distribution")
```

The next graph depicts the seaborn kernel density plot of authors and their top 20 TF-IDF ranked words used. It is interesting to see that there is almost a normal distribution of the TF-IDF among some authors. We can try to naively predict a text's author by calculating the TF-IDF on the unknown text and comparing it to each author's distribution. Authors may tend to use a certain set of relevant words in all of their writings.

```python
plt.figure(figsize=(10,10))    
for author in author_names:
    tfidf_mat = tfidf_vect.fit_transform(by_author[author])
    df = pd.DataFrame(data=tfidf_mat.toarray(),columns=tfidf_vect.get_feature_names())
    most_rel = df.max().sort_values()[-20:]
    sns.kdeplot(np.array(most_rel), color=next(palette), shade=True, label=author)
    
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Author's Most Relevant Word Distribution")
plt.show()
```

The following graph is a visualization of each of 8 authors and how frequently they use their top 20 most common words.
```python
author_names = ['AaronPressman', 'AlanCrosby', 'AlexanderSmith', 'BenjaminKangLim', 'BernardHickey',
                'BradDorfman', 'DarrenSchuettler']
plt.figure(figsize=(10,10))    

for author in author_names:
    bag_words = count_vect.fit_transform(by_author[author])
    df_frequency = pd.DataFrame(data=bag_words.toarray(),columns=count_vect.get_feature_names())

    plt.plot(df_frequency.sum().sort_values()[-20:].values, label=author)

    
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Author's Most Frequent Words")
plt.xlabel("Least-Most Frequent Word")
plt.ylabel("Frequency")
plt.show()
```
